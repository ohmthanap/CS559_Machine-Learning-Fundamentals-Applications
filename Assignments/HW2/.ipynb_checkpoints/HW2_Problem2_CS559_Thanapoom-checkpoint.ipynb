## Problem 2
Face modelling serves as one of the most fundamental problems in modern artificial intelligence and computer vision, and can be useful in various applications like face recognition, identification etc. However, face images are usually of high-dimensional (e.g., a small 100×100 gray-scaled face image has dimension 100 × 100 = 10, 000), therefore, find a suitable representation is utterly important. In this problem, we apply the linear model, principal component analysis (PCA), on face images to reduce the dimension and obtain eigenface representations.

**Dataset**: we use the dataset which contains 177 face images. Each image contains 256 × 256 pixels and is gray-scaled (i.e., the value for each pixel is stored as unsigned integer between [0, 255], typically, 0 is taken to be black and 255 is taken to be white). You need to split the dataset to be train/test set, e.g., you could use the first 157 images for training, and the rest 20 faces for testing.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import glob
from PIL import Image
```

```python
# Import the image data
# Define the path of the folder contains the images
images_folder = "/Users/thanapoomphatthanaphan/Education/Master Degree/Courses/CS559_Machine-Learning-Fundamentals-Applications/Assignments/HW2/face_data/*.bmp"

# Get a list, containing the path of each image
image_path = glob.glob(images_folder)

# Load the image objects, and put them in the list
images_data = []
for image in image_path:
    images_data.append(np.array(Image.open(image)))
```

**(1)** [30pt] Write the PCA codes to compute K = 30 eigenfaces and visualize the top 10 eigenfaces.

```python
# Reshape the matrix to 1D
def reshape_1d(images_data):
    
    """A function to reshape the matrix to 1D"""
    
    # Reshape each image from 2D to 1D vector
    images_1d = []
    for image in images_data:
        images_1d.append(image.flatten())
    images_1d = np.array(images_1d)

    return images_1d
    
# Get the mean images
def compute_mean_image(images_data):
    
    """A function to get the mean image"""
    
    # Reshape the images data into 1d
    images_1d = reshape_1d(images_data)
    
    # Compute the mean image
    n = len(images_1d)
    mean_image = np.sum(images_1d) / n
    
    return mean_image
    
# Perform PCA
def perform_pca(images_data, K, mean_image):
    
    """A function to perform PCA to get the top K eigenfaces"""
    
    # Reshape each image from 2D to 1D vector
    images_1d = reshape_1d(images_data)
    
    # Recenter each training image by subtracting with the mean image
    images_centered = images_1d - mean_image

    # Step 5: Perform SVD
    U, S, Vt = np.linalg.svd(images_centered, full_matrices=False)
    
    # Get the top K eigenfaces
    top_K_eigenfaces = Vt[:K]
    
    return top_K_eigenfaces

# Define the top K
K = 30

# Get the mean image of training images
mean_image = compute_mean_image(images_data[:157])

# Call the function to perform PCA and visualize eigenfaces
top_K_eigenfaces = perform_pca(images_data[:157], K, mean_image)

# Visualize the top 10 eigenfaces
for i in range(10):
    plt.figure(figsize=(2, 2))
    plt.imshow(top_K_eigenfaces[i, :].reshape(256, 256), cmap='gray')
    plt.title(f'Eigenface {i + 1}')
    plt.axis('off')
    plt.show()
```

**(2)** [20pt] Use the learned $K$ eigenfaces from (1) to reconstruct testing images, and record the reconstruction error. The reconstruction error can be defined as  $\dfrac{||\hat Y − Y||^2}{N}$ , where $\hat Y$ is the reconstructed face using the learned eigenfaces, $Y$ is the testing faces and $N$ is the total number of testing data. Please show 5 testing images and their reconstructed ones.

```python
def reconstruct_images(testing_data, mean_image, top_K_eigenfaces):
        
        """A function to reconstruct images"""
        
        # Reshape each image from 2D to 1D vector
        images_1d = reshape_1d(testing_data)
        
        # Recenter each testing image by subtracting with the mean image
        testing_images_centered = images_1d - mean_image
        
        # Project testing images onto the K eigenfaces
        projected_testing_images = np.dot(testing_images_centered, top_K_eigenfaces.T)
        
        # Approximate the centered testing image using K eigenfaces
        reconstructed_testing_images = np.dot(projected_testing_images, top_K_eigenfaces)
        
        # Add back the mean image to obtain the reconstrution for testing image
        reconstructed_testing_images += mean_image

        # Normalize the testing images and reconstructed images for computing the reconstruction error
        testing_images_centered_norm = (testing_images_centered - np.mean(testing_images_centered)) / np.std(testing_images_centered)
        reconstructed_testing_images_norm = (reconstructed_testing_images - np.mean(reconstructed_testing_images)) / np.std(reconstructed_testing_images)
        
        # Calculate the reconstruction error
        reconstruction_error = np.mean((reconstructed_testing_images_norm - testing_images_centered_norm) ** 2)
        print("Reconstruction_error at K =", K, ":", reconstruction_error)
        
        return reconstructed_testing_images, reconstruction_error

reconstructed_testing_images, recon_error = reconstruct_images(images_data[-20:], mean_image, top_K_eigenfaces)

# Convert each image from 2D to 1D
original_images = reshape_1d(images_data[-20:])

# Display 5 testing images and their reconstructed counterparts
for i in range(5):
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(original_images[i].reshape(256, 256), cmap='gray')
    plt.title('Original')
    plt.subplot(1, 2, 2)
    plt.imshow(reconstructed_testing_images[i].reshape(256, 256), cmap='gray')
    plt.title('Reconstructed')
```

**(3)** [20pt] Try different values of *K* , e.g., try *K* = 10, 30, 50, 100, 150..., and draw the curve to indicate the corresponding testing reconstruction errors. The x-axis of the curve can be different *K* values, and the y-axis can be testing reconstruction error defined in (2).

```python
# Different values of K for testing
K_values = [10, 30, 50, 100, 150]
errors = []

# Computing reconstruction errors at each K value
for K in K_values:
    print("Testing at K:", K)
    top_K_eigenfaces = perform_pca(images_data[:157], K, mean_image)
    reconstructed_testing_images, recon_error = reconstruct_images(images_data[-20:], mean_image, top_K_eigenfaces)
    errors.append(recon_error)
    
# Plot the errors
plt.plot(K_values, errors, marker='o')
plt.xlabel('K (Number of Eigenfaces)')
plt.ylabel('Testing Reconstruction Error')
plt.title('Testing Reconstruction Error vs. K')
plt.grid(True)
plt.show()

# Display 5 testing images and their reconstructed counterparts at K = 150
print("5 Testing images and their reconstructed counterparts at K = 150")
for i in range(5):
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(original_images[i].reshape(256, 256), cmap='gray')
    plt.title('Original')
    plt.subplot(1, 2, 2)
    plt.imshow(reconstructed_testing_images[i].reshape(256, 256), cmap='gray')
    plt.title('Reconstructed')
```
